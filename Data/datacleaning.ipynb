```python
import pandas as pd
```

```python

# Load dataset
df = pd.read_csv("Instagram data.csv", encoding="latin1")

# Clean column names
df.columns = [c.strip().lower().replace(" ", "_") for c in df.columns]

```

```python
# Create engagement metrics
df["total_engagement"] = (
    df["likes"] + df["comments"] + df["shares"] + df["saves"]
)

df["engagement_rate"] = df["total_engagement"] / df["impressions"]
```

```python
# Text-based features
df["hashtag_count"] = (
    df["hashtags"]
    .fillna("")
    .astype(str)
    .apply(lambda x: sum(word.startswith("#") for word in x.split()))
)
df["caption_length"] = df["caption"].fillna("").apply(len)
```

```python
df["passive_engagement"] = df["likes"]

df["active_engagement"] = (
    df["comments"] + df["shares"] + df["saves"]
)

df["active_engagement_per_1k_impressions"] = (
    df["active_engagement"] / df["impressions"]
) * 1000
```

```python
# Save processed data
df.to_csv("instagram_processed.csv", index=False)

print("Data processing complete!")
```
